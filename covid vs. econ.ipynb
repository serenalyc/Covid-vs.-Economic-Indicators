```python
missing_values = "."
for i in range(16642):
    if data.loc[i]['new_case_rate'] == missing_values or data.loc[i]['new_death_rate'] == missing_values or data.loc[i]["new_case_count"] == missing_values or data.loc[i]["new_death_count"]==missing_values:
        data = data.drop(i)
data
```

```python
group_by_month = data.groupby(['month']).mean()
group_by_month.drop(['day','cityid'],axis=1)
```

```python
import numpy as np
months=np.arange(1,12)
```

```python
import matplotlib.pyplot as plt
dataset_1 = group_by_month['case_count']
dataset_2 = group_by_month['death_count']
dataset_3 = group_by_month['case_rate']
dataset_4 = group_by_month['death_rate']

fig = plt.figure(figsize=(10,10))
axes1 = fig.add_subplot(2,2,1)
axes1.set_xticks(months)
axes2 = fig.add_subplot(2,2,2)
axes2.set_xticks(months)
axes3 = fig.add_subplot(2,2,3)
axes3.set_xticks(months)
axes4 = fig.add_subplot(2,2,4)
axes4.set_xticks(months)

axes1.plot(months,dataset_1)
axes2.plot(months,dataset_2)
axes3.plot(months,dataset_3)
axes4.plot(months,dataset_4)

fig.suptitle("Mean Covid Growth Cases/Rates for Each Month")

axes1.set_title("Case Count",color = 'r')
axes2.set_title("Death Count",color = 'r')
axes3.set_title("Case Rate",color = 'r')
axes4.set_title("Death Rate",color = 'r')

axes1.set_xlabel("Month")
axes1.set_ylabel("Case counted")
axes2.set_xlabel("Month")
axes2.set_ylabel("Death Counted")
axes3.set_xlabel("Month")
axes3.set_ylabel("Case Rate")
axes4.set_xlabel("Month")
axes4.set_ylabel("Deah Rate")
```
